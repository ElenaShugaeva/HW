```python
# Определите какому фильму было выставлено больше всего оценок 5.0.
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
ratings[ ratings.rating == 5.0 ].groupby('movieId').count().sort_values('rating', ascending=False).\
merge(movies, on='movieId', how='left').head()
```

```python
# По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21
# за период с 2005 по 2010 года.
# Не учитывайте в расчетах отрицательные значения quantity.
```

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
filtered_countries = data[ ((data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia'))
     & (data.year < 2011) & (data.year > 2004) & ((data.category == 4) | (data.category == 12) | 
                                                  (data.category == 21)) & (data.quantity >= 0)]
filtered_countries.head(10)
```

```python
filtered_countries['quantity'].sum()
```
