```python
# Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, 
# которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением
# столбца timestamp для данного значения userId.
```

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings1 = ratings.groupby('userId').agg({'timestamp': ['min', 'max'], 'rating':'count'})
ratings1['timestamp_diff'] =  ratings1['timestamp']['max']-ratings1['timestamp']['min']
ratings1.head()
```

```python
ratings1.loc[ratings1['rating']['count'] > 100].head()
```

```python
ratings1.timestamp_diff.mean()/ 24 / 3600
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table = pd.concat([rzd, auto, air, client_base])
table.groupby('client_id').agg ({'air_revenue':'sum', 'rzd_revenue': 'sum', 'auto_revenue':'sum', 'address': 'last'})
```

```python
table.pivot_table(index = 'client_id', fill_value = 0)
```
