```python
import pandas as pd
import numpy as np
```

```python
horse_df = pd.read_csv('horse_data.csv')
horse_df = pd.read_csv('horse_data.csv', na_values = "?")
horse_df.head()
```

```python
horse_df.rename(columns={ horse_df.columns[1]: "Age", horse_df.columns[2]: "Hospital_number", horse_df.columns[4]: "Pulse", \
                       horse_df.columns[6]: "Temp_of_extremities", horse_df.columns[7]: "Peripheral_pulse", horse_df.columns[11]: "Peristalsis",\
                       horse_df.columns[12]: "Abdominal_distension", horse_df.columns[22]: "Outcome"  }, inplace=True)
hs_df = horse_df[['Age', 'Hospital_number', 'Pulse','Temp_of_extremities','Peripheral_pulse', 'Peristalsis','Abdominal_distension','Outcome']]
hs_df.head()
```

```python
for col in hs_df.columns:
    missing = hs_df[col].isnull().mean()
    print(f'{col} - {missing :.1%}')
```

```python
# процент пропусков по температуре конечностей обЪясним- является субъективной характеристикой- пропуски носят случайный характер 
# - и скорее всего вызваны состоянием, близким к нормальному - пропуски принимаются за константу нормального состояния = 1
# Аналогично = пропуски перифирического пульса принимаются за 1, перистальтики = 2.
# Пульс - характеристика точно измеряемая и пропуски малообЪяснимы, как показатели используются наиболее массовые результаты 
# - мода.
# Вздутие живота- важный показатель, поэтому отсутствие его малообяъснимо , случайно, в идеале должно браться в зависимости от 
# комбинации нескольких параметров. Здесь для упрощения используется мода.
# Пропуски результата лечения = игнорируются.
```

```python
hs_df.Pulse.fillna(hs_df.Pulse.mode()[0], inplace=True)
hs_df.Abdominal_distension.fillna(hs_df.Abdominal_distension.mode()[0], inplace=True)
hs_df.Temp_of_extremities.fillna(1.0, inplace = True)
hs_df.Peripheral_pulse.fillna(1.0, inplace = True)
hs_df.Peristalsis.fillna(2.0, inplace = True)
hs_df.head()
```
