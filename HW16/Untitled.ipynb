```python
titles = pd.read_csv('netflix_titles.csv')
titles.head()
```

```python
titles.info()
```

```python
grouped1 = titles.groupby(['rating']).count()
grouped1
```

```python
plt.plot(grouped1.index, grouped1.show_id)
plt.title('   ')
plt.xlabel('')
plt.ylabel(' ')
```

```python
grouped2 = titles.groupby(['director']).count()
grouped1.loc[['TV-14', 'TV-MA']].groupby(['director']).count()
```

```python
grouped3 = titles[titles['rating']. isin (['TV-14', 'TV-MA'])]
grouped3.head()
```

```python
grouped4 = grouped3.groupby(['director']).count()
grouped4
```

```python
grouped5= grouped4.sort_values('rating', ascending=False)[:50]
grouped5.head()
```

```python
plt.plot(grouped5.index, grouped5.rating 
#                      bins=11,
#                      title='111',
#                      xlabel='director',
#                      ylabel='show_id',
#                      xticks=range(1, 11))
        )
```
